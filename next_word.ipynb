```python
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
```

```python
data=gutenberg.raw('shakespeare-hamlet.txt')
```

```python
with open('hamlet.txt','w') as f:
  f.write(data)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
with open('hamlet.txt','r') as f:
  data=f.read().lower()
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
print(tokenizer.word_counts)
```

```python
total_words=len(tokenizer.word_index)+1
total_words
```

```python
print(tokenizer.word_index)
```

```python
 input_s=[]
 for i in data.split('\n'):

  temp=tokenizer.texts_to_sequences([i])[0]
  for j in range(1,len(temp)):
    input_s.append(temp[:j+1])
```

```python
max_seq=max([len(i) for i in input_s])
```

```python
input_seq=np.array(pad_sequences(input_s,maxlen=max_seq,padding='pre'))
```

```python
x,y=input_seq[:,:-1],input_seq[:,-1]
```

```python
y[0]
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
```

```python
# First, verify these values
print("Total words:", total_words)
print("Max sequence length:", max_seq)

# Then build the model with explicit input shape
model=Sequential()
model.add(Embedding(input_dim=total_words, output_dim=100, input_length=max_seq-1))
model.add(LSTM(150, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

# Compile the model
#model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Build the model by either:
# Method 1: Using build()
model.build(input_shape=(None, max_seq-1))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(x_train,y_train,epochs=200,validation_data=(x_test,y_test))
```

```python
model=model.save('mymodel.h5')
```
